```python
# The five plots of seaborn library are:- 
# 1.replot 
import seaborn as sns 
import pandas as pd 
datum = pd.read_csv("student.csv")
df = sns.relplot(datum) 
# It uses scattering technique to plot the different data sets 
# in a file
```

```python
# 2.scatterplot 

df2 = sns.lineplot(datum)
df2 
# It plots graph by making connecting lines between the points
```

```python
# 3.boxplot 
df3 = sns.boxplot(datum)
df3 
# It plots graph based on box method
```

```python
# 4.barplot


sns.barplot(datum)
#  Bar plots are useful for comparing the mean or aggregated values of different categories. They are commonly used to show the distribution of a categorical variable.
```

```python
# 5.pairplot 

sns.pairplot(datum) 

# pairplot is used to make graph by relation of all the columns
# present in the data
```

```python


# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(flights_pivot)

# Set plot title
plt.title('Passenger Count by Month and Year')

# Show the plot
plt.show()
```
