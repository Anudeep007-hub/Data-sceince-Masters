Q-1

Matplotlib is a python tool used for making graphs, animated, static graphs. It has multiple packages for creating piecharts, bardiagram, histogram, even implementing in 3d graphs. With these one can use it for creating various projects, applications, etc., 


```python
# Five plots that can be plotted using Pyplot,
# 1 - plot 

import matplotlib.pyplot as plt 
import numpy as np 

x = np.random.randint(-12,20,23)
y = np.random.randint(-18,30,23)
plt.plot(x,y)
```

```python
# 2 - stem 
plt.stem(x,y)
```

```python
# 3 - piechart 


labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 90, 20, 25]

plt.pie(sizes, labels=labels, autopct="%1f%%" ,startangle=40)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title('Pie Chart')



```

```python
# 4 - bar graph 
plt.bar(x,y)
plt.grid()
```

```python
# 5 - histograph  
import numpy as np
m = list(np.random.randint(1,28,30))
plt.hist(m)
plt.grid()
plt.show()
```

Q-2

A scatter plot is a type of plot that displays individual data points on a two-dimensional plane, with each point represented by a marker.

```python
import numpy as np
import matplotlib.pyplot as plt 
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y, c="#211e1e") 
plt.xlabel("X->")
plt.ylabel("Y->") 
plt.title("Graph") 
plt.show()
```

Q-3

subplot is used for fititng  multiple graphs in a single page. Generally it is used for better comparing the tend of graphs.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

plt.tight_layout() 

plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays key statistical measures such as the median, quartiles, and potential outliers. The box plot consists of a rectangular "box" and two "whiskers" that extend from the box.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200) 
m = [box1,box2]
plt.boxplot(m, labels = ["Box1", "Box2"])
plt.show()
```
